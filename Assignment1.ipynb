5.Visualisations using the dataset

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

#COUNTPLOT

```python
sns.countplot(x='Classification', data=data)
plt.title('Countplot of Iris Species')
plt.xlabel('Classification')
plt.ylabel('Count')
plt.show()
```

#BOX PLOT

```python
plt.title('Boxplot of Iris Features by Classification')
sns.boxplot(x='Classification', y='SW', data=data)
plt.show()
```

SCATTER PLOT

```python
sns.scatterplot(x='SL', y='SW', hue='Classification', data=data, edgecolor='b', s=150)
plt.title('Scatter Plot of SL vs SW by Classification')
plt.xlabel('SL')
plt.ylabel('SW')
plt.show()
```
